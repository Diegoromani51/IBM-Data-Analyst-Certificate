```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
mpl.style.use('ggplot')

QUERY = """
SELECT *
FROM master"""

df_hist = pd.read_sql_query(QUERY, conn)
df_hist['ConvertedComp'].plot(kind='hist')
plt.title('Converted Compensation Histogram')
plt.xlabel("Annual USD")
plt.ylabel('Number of Respondents')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

QUERY = """
SELECT *
FROM master """

df_box = pd.read_sql_query(QUERY, conn)
df_box['Age'].plot(kind='box', figsize=(10,8))
plt.title('Box plot of Respondents Age')
plt.ylabel('Age Ranges')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

QUERY = """
SELECT Age, WorkWeekHrs
FROM master """

df_scatter = pd.read_sql_query(QUERY, conn)
sns.scatterplot(x='Age', y='WorkWeekHrs', data=df_scatter)
plt.title('Relation Between Age vs. WorkWeekHrs for Respondents')
plt.xlabel('Age of Respondents')
plt.ylabel('Work Week Hours')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master """

df_bubble = pd.read_sql_query(QUERY, conn)
sns.scatterplot(data= df_bubble, x='WorkWeekHrs', y='CodeRevHrs', size='Age', hue='Age', alpha=0.5, sizes=(10,400))
plt.title('Work Week Hours vs. Code Rev Hours by Age of Respondents')
plt.xlabel('Work Week Hours')
plt.ylabel('Code Rev Hours')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

QUERY = """
SELECT *
FROM DatabaseDesireNextYear """

df_db = pd.read_sql_query(QUERY, conn)
df_db = df_db.groupby('DatabaseDesireNextYear', axis=0).count()
df_db.sort_values(['Respondent'], ascending=False, inplace=True)
df_pie = df_db.head()

df_pie['Respondent'].plot(kind='pie',
                         figsize=(5,6),
                         autopct='%1.1f%%',
                         startangle=90,
                         labels=df_db.index
                         )
plt.title('Top 5 Databases that Respondents Wish to Learn Next Year')
plt.axis('equal')
plt.show()
```

```python
QUERY = """
SELECT *
FROM LanguageDesireNextYear """

df_lg = pd.read_sql_query(QUERY, conn)
df_lg = df_lg.groupby('LanguageDesireNextYear', axis=0).count()
df_lg.sort_values(['Respondent'], ascending=False, inplace=True)
df_lg
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
From master
WHERE Age between 30 AND 35 """

df_stack = pd.read_sql_query(QUERY, conn)

df_stack = df_stack.groupby(['CodeRevHrs','Age'])['WorkWeekHrs'].median().unstack()
df_stack.plot(kind='bar', stacked=True, figsize=(25,10))
plt.title('Stacked Chart of Median Work Week Hours and Code Rev Hours for The Age Group 30 to 35', loc='center', fontsize=25)
plt.ylabel('Work Week Hours', fontsize=20)
plt.xlabel('CodeRev Hours', fontsize=20)
plt.xticks(rotation=0)
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

QUERY = """
SELECT ConvertedComp, Age
FROM master
Where Age between 45 AND 60 """

df_line = pd.read_sql_query(QUERY, conn)
df_line = df_line.groupby('Age').median()

df_line.plot(kind='line')
plt.title('Median Converted COmpensation for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median Salary')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

QUERY = """
SELECT MainBranch, COUNT(*) as Number_of_Respondents
FROM master
group by MainBranch """

    
df_bar = pd.read_sql_query(QUERY, conn)

df_bar.plot(kind='barh')
plt.yticks(df_bar.index, labels=df_bar['MainBranch'])
plt.title('Total Respondents Based on Main Branch')
plt.xlabel('Total')

plt.show()
df_bar
```
